```python
import os
from io import StringIO

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/9_3_h_clustering_c5_with_structure.csv')
print(len(df))
df.head(5)
```

```python
sliced_df = np.array_split(df,9)
sliced_df[0]
```

```python
sliced_df = np.array_split(df,9)
sliced_df[0]
```

```python

for n,d in enumerate(sliced_df):
    first_folder = './'+str(n+1)
    os.mkdir(first_folder)
    
    for i in range(len(d)):
        s = d.iloc[i,:]
        index = s['Unnamed: 0']
        mpid = s['mpid']
        composition = s['composition']
        folder_name = first_folder+'/'+composition+'_'+mpid+'_'+ str(index)
        os.mkdir(folder_name)

        with open(folder_name +'/input.POSCAR','w') as p:
            p.write(s['structure_poscar'])
        with open(folder_name +'/info.txt','w') as p:
            p.write('Predicted results\n\n')
            p.write(str(s.keys))
```
